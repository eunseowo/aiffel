### 3-1. ë°ì´í„° ì „ì²˜ë¦¬ [ë²¡í„°í™” : tf-idf]

```python
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=10000, test_split=0.2)
```

```python
word_index = reuters.get_word_index(path="reuters_word_index.json")
```

```python
index_to_word = { index+3 : word for word, index in word_index.items() }
for index, token in enumerate(("<pad>", "<sos>", "<unk>")):
  index_to_word[index]=token
```

```python
decoded = []
for i in range(len(x_train)):
    t = ' '.join([index_to_word[index] for index in x_train[i]])
    decoded.append(t)

x_train = decoded
print(len(x_train))
```

```python
decoded_test = []
for i in range(len(x_test)):
    t = ' '.join([index_to_word[index] for index in x_test[i]])
    decoded_test.append(t)

x_test = decoded_test
print(len(x_test))
```

```python
# ë²¡í„°í™” DTM, TF-idf ë°©ë²•
dtmvector = CountVectorizer()

tfidf_transformer = TfidfTransformer()

x_train_dtm = dtmvector.fit_transform(x_train)
x_test_dtm= dtmvector.transform(x_test)

x_train_tfidf = tfidf_transformer.fit_transform(x_train_dtm)
x_test_tfidf = tfidf_transformer.transform(x_test_dtm)
```

#### 3-2-1. ComplementNB (CNB)

```python
from sklearn.naive_bayes import ComplementNB
```

```python
cb = ComplementNB()
cb.fit(x_train_tfidf, y_train)
```

```python
# ì˜ˆì¸¡
y_pred = cb.predict(x_test_tfidf)

# í‰ê°€ ì§€í‘œ
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"âœ… Accuracy : {acc:.4f}")
print(f"âœ… F1-score : {f1:.4f}")
```

### 4-1. ë°ì´í„° ì „ì²˜ë¦¬ [ë²¡í„°í™” : tf-idf, W2V]

- ë‹¨ì–´ ì œí•œ ìˆ˜ ì—†ì´ ì „ì²´ ë‹¨ì–´ë¥¼ ì‚¬ìš©í–ˆì„ ë•Œ ê°€ì¥ ì¢‹ì€ ì„±ëŠ¥ì„ ë³´ì„.
    - `num_words = None` ìœ¼ë¡œ ì§„í–‰

```python
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=None, test_split=0.2)
```

```python
word_index = reuters.get_word_index(path="reuters_word_index.json")
```

```python
index_to_word = { index+3 : word for word, index in word_index.items() }
for index, token in enumerate(("<pad>", "<sos>", "<unk>")):
  index_to_word[index]=token
```

```python
decoded = []
for i in range(len(x_train)):
    t = ' '.join([index_to_word[index] for index in x_train[i]])
    decoded.append(t)

x_train = decoded
print(len(x_train))
```

```python
decoded_test = []
for i in range(len(x_test)):
    t = ' '.join([index_to_word[index] for index in x_test[i]])
    decoded_test.append(t)

x_test = decoded_test
print(len(x_test))
```

```python
# ë²¡í„°í™” DTM, TF-idf ë°©ë²•
dtmvector = CountVectorizer()

tfidf_transformer = TfidfTransformer()

x_train_dtm = dtmvector.fit_transform(x_train)
x_test_dtm= dtmvector.transform(x_test)

x_train_tfidf = tfidf_transformer.fit_transform(x_train_dtm)
x_test_tfidf = tfidf_transformer.transform(x_test_dtm)
```

```python
x_train[3]
```

```python
# ë²¡í„°í™” W2Vë°©ë²•
from gensim.models import Word2Vec

# ìš°ì„  ë¬¸ì¥ì„ í† í°í™” ì‹œí‚µì‹œë‹¤ ë„ì–´ì“°ê¸° ê¸°ë°˜ìœ¼ë¡œ í•´ë³¼ê²Œìš”! -> # ìœ„ì—ì„œ DTMë§Œë“¤ë•ŒëŠ” ì™œ ì•ˆí•´ì¤¬ëƒ! -> CountVectorizerì—ì„œ ë„ì–´ì“°ê¸° ê¸°ë°˜ í† í°í™”ê°€ ë‚´ì¥ë˜ìˆìŒ
x_train_tokenized = [sentence.split() for sentence in x_train]
x_test_tokenized = [sentence.split() for sentence in x_test]

# vectorì‚¬ì´ì¦ˆë¥¼ ëŠ˜ë¦¬ê±°ë‚˜ ì¤„ì—¬ë³´ì„¸ìš” ì•„ë§ˆ 512 ê°€ì¥ë§ì´ì“°ì´ëŠ” ë°©ì‹
model = Word2Vec(sentences = x_train_tokenized, vector_size = 256, window = 5, min_count = 5, workers = 4, sg = 0)
print("ëª¨ë¸ í•™ìŠµ ì™„ë£Œ!")
```

```python
# W2Vì´ ì˜ë˜ì—ˆëŠ”ì§€ í™•ì¸ -> ì—¬ì°¨ì €ì°¨ ë˜ê¸´í•œê²ƒê°™ë‹¤
model_result = model.wv.most_similar('man')
print(model_result)
```

```python
# í•™ìŠµëœ Word2Vec ëª¨ë¸
w2v_model = model

# ê° ë¬¸ì¥ì„ ë²¡í„°í™” ì‹œí‚¤ëŠ” ì½”ë“œ
def vectorize_sentence(sentence, model, max_len):
    vecs = []
    for word in sentence:
        if word in model.wv:
            vecs.append(model.wv[word])
        else:
            vecs.append(np.zeros(model.vector_size))
    # Padding
    if len(vecs) < max_len:
        vecs += [np.zeros(model.vector_size)] * (max_len - len(vecs))
    else:
        vecs = vecs[:max_len]
    return np.array(vecs)


# ìµœëŒ€ ë¬¸ì¥ê¸¸ì´ë¥¼ ì˜ ì¡ì•„ì£¼ì„¸ìš”
x_train_w2v = np.array([vectorize_sentence(s, w2v_model, max_len=100) for s in x_train_tokenized])
x_test_w2v = np.array([vectorize_sentence(s, w2v_model, max_len=100) for s in x_test_tokenized])
```
