## 3. 빈도수 상위 10,000개 단어 사용

### 3-1. 데이터 전처리 [벡터화 : tf-idf]

```python
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=10000, test_split=0.2)
```

```python
word_index = reuters.get_word_index(path="reuters_word_index.json")
```

```python
index_to_word = { index+3 : word for word, index in word_index.items() }
for index, token in enumerate(("<pad>", "<sos>", "<unk>")):
  index_to_word[index]=token
```

```python
decoded = []
for i in range(len(x_train)):
    t = ' '.join([index_to_word[index] for index in x_train[i]])
    decoded.append(t)

x_train = decoded
print(len(x_train))
```

```python
decoded_test = []
for i in range(len(x_test)):
    t = ' '.join([index_to_word[index] for index in x_test[i]])
    decoded_test.append(t)

x_test = decoded_test
print(len(x_test))
```

```python
# 벡터화 DTM, TF-idf 방법
dtmvector = CountVectorizer()

tfidf_transformer = TfidfTransformer()

x_train_dtm = dtmvector.fit_transform(x_train)
x_test_dtm= dtmvector.transform(x_test)

x_train_tfidf = tfidf_transformer.fit_transform(x_train_dtm)
x_test_tfidf = tfidf_transformer.transform(x_test_dtm)
```

```python
from sklearn.naive_bayes import ComplementNB
```

```python
cb = ComplementNB()
cb.fit(x_train_tfidf, y_train)
```

```python
# 예측
y_pred = cb.predict(x_test_tfidf)

# 평가 지표
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"✅ Accuracy : {acc:.4f}")
print(f"✅ F1-score : {f1:.4f}")
```

### 4-1. 데이터 전처리 [벡터화 : tf-idf, W2V]

- 단어 제한 수 없이 전체 단어를 사용했을 때 가장 좋은 성능을 보임.
    - `num_words = None` 으로 진행

```python
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=None, test_split=0.2)
```

```python
word_index = reuters.get_word_index(path="reuters_word_index.json")
```

```python
index_to_word = { index+3 : word for word, index in word_index.items() }
for index, token in enumerate(("<pad>", "<sos>", "<unk>")):
  index_to_word[index]=token
```

```python
decoded = []
for i in range(len(x_train)):
    t = ' '.join([index_to_word[index] for index in x_train[i]])
    decoded.append(t)

x_train = decoded
print(len(x_train))
```

```python
decoded_test = []
for i in range(len(x_test)):
    t = ' '.join([index_to_word[index] for index in x_test[i]])
    decoded_test.append(t)

x_test = decoded_test
print(len(x_test))
```

```python
# 벡터화 DTM, TF-idf 방법
dtmvector = CountVectorizer()

tfidf_transformer = TfidfTransformer()

x_train_dtm = dtmvector.fit_transform(x_train)
x_test_dtm= dtmvector.transform(x_test)

x_train_tfidf = tfidf_transformer.fit_transform(x_train_dtm)
x_test_tfidf = tfidf_transformer.transform(x_test_dtm)
```

```python
x_train[3]
```

```python
# 벡터화 W2V방법
from gensim.models import Word2Vec

# 우선 문장을 토큰화 시킵시다 띄어쓰기 기반으로 해볼게요! -> # 위에서 DTM만들때는 왜 안해줬냐! -> CountVectorizer에서 띄어쓰기 기반 토큰화가 내장되있음
x_train_tokenized = [sentence.split() for sentence in x_train]
x_test_tokenized = [sentence.split() for sentence in x_test]

# vector사이즈를 늘리거나 줄여보세요 아마 512 가장많이쓰이는 방식
model = Word2Vec(sentences = x_train_tokenized, vector_size = 256, window = 5, min_count = 5, workers = 4, sg = 0)
print("모델 학습 완료!")
```

```python
# W2V이 잘되었는지 확인 -> 여차저차 되긴한것같다
model_result = model.wv.most_similar('man')
print(model_result)
```

```python
# 학습된 Word2Vec 모델
w2v_model = model

# 각 문장을 벡터화 시키는 코드
def vectorize_sentence(sentence, model, max_len):
    vecs = []
    for word in sentence:
        if word in model.wv:
            vecs.append(model.wv[word])
        else:
            vecs.append(np.zeros(model.vector_size))
    # Padding
    if len(vecs) < max_len:
        vecs += [np.zeros(model.vector_size)] * (max_len - len(vecs))
    else:
        vecs = vecs[:max_len]
    return np.array(vecs)


# 최대 문장길이를 잘 잡아주세요
x_train_w2v = np.array([vectorize_sentence(s, w2v_model, max_len=100) for s in x_train_tokenized])
x_test_w2v = np.array([vectorize_sentence(s, w2v_model, max_len=100) for s in x_test_tokenized])
```

### 4-2. 머신러닝 vs 딥러닝

- 동일한 데이터셋과 전처리 조건 (Voca_Size = None)으로 딥러닝 모델의 성능과 비교하자.
- 가장 좋았던 모델인 `LogisticRegression` 으로 진행

#### 4-2-1. LogisticRegression + tf-idf

```python
lr = LogisticRegression(C=10000, penalty='l2', max_iter=3000)
lr.fit(x_train_tfidf, y_train)
```

```python
# 예측
y_pred = lr.predict(x_test_tfidf)

# 평가 지표
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"✅ Accuracy : {acc:.4f}")
print(f"✅ F1-score : {f1:.4f}")
```

#### 4-2-2. LogisticRegression + W2V

```python
# 데이터를 단어단위에서 문장단위로 바꿔줘야합니다.. ML은 2차원데이터만 받을수있기때문
# 문장에 대해서 토큰들의 벡터를 평균을 취해줍니다.

# Word2Vec 임베딩 시퀀스: (8982, 100, 256)
x_w2v_seq_train = x_train_w2v
x_w2v_seq_test = x_test_w2v
# 평균 풀링 → (8982, 256)
x_w2v_avg_train = np.mean(x_w2v_seq_train, axis=1)
x_w2v_avg_test = np.mean(x_w2v_seq_test, axis=1)
print(x_w2v_avg_train.shape)  # (8982, 256)
```

```python
lr = LogisticRegression(C=10000, penalty='l2', max_iter=3000)
lr.fit(x_w2v_avg_train, y_train)
```

- 시간이 너무 오래 걸린다...
- 이유?
    - Reuters 데이터셋은 클래스가 46개로 많음.
    - LogisticRegression은 기본적으로 One-vs-Rest (OvR) 전략
        - 이진 분류기 46개를 학습해야함.
        - 즉, 내부적으로는 46개의 회귀 문제를 병렬로 풀고 있음 → 시간이 오래 걸릴 수 있음

 

```python
# 예측
y_pred = lr.predict(x_w2v_avg_test)

# 평가 지표
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"✅ Accuracy : {acc:.4f}")
print(f"✅ F1-score : {f1:.4f}")
```

#### 4-2-3. 벡터화 방법별 ML 모델 성능 비교

**성능 비교**
| Vectorization | Model              | Accuracy   | F1-Score   |
| ------------- | ------------------ | ---------- | ---------- |
| **TF-IDF**    | LogisticRegression | **0.8112** | **0.8055** |
| **Word2Vec**  | LogisticRegression | 0.7333     | 0.7331     |

**왜 이렇게 나왔을까?**

| 비교 항목                   | TF-IDF                         | Word2Vec 평균                              |
| ----------------------- | ------------------------------ | ---------------------------------------- |
| 표현 방식                   | 단어 빈도 기반 (희소 벡터)               | 의미 기반 임베딩 (밀집 벡터)                        |
| 장점                      | 각 단어의 중요도를 반영하여 **정확하게 분리 가능** | 문장 전체 의미를 반영하지만 **세밀한 분류 경계가 흐려짐**       |
| LogisticRegression과의 궁합 | 선형 분류기에 잘 맞는 **고차원 희소 구조**     | 의미 표현은 가능하지만 선형 결정 경계에선 **분류 성능 손실** 가능성 |

**결론**
- `LogisticRegression`은 TF-IDF처럼 고차원이고 희소한 특성에서 더 나은 분류 성능을 보이며,
Word2Vec 평균은 문장 의미를 담긴 했지만, 클래스 간 경계를 명확히 하기엔 정보가 압축되어 있었음.


#### 4-2-4. Dense NN + W2V

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout, LSTM, Dense, Dropout


dense_model = Sequential([
    Flatten(input_shape=(100, 256)),  # (seq_len, embedding_dim)
    Dense(512, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(46, activation='softmax')   # 클래스 수에 맞게 조정 46개로 맞춰주세요!
])

dense_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
dense_model.summary()
```

```python
# rnn 시계열 특징 데이터 특화 모델

rnn_model = Sequential([
    LSTM(128, input_shape=(100, 256)),  # (seq_len, embedding_dim)
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(46, activation='softmax')   # 클래스 수에 맞게 조정 46개로 맞춰주세요~
])

rnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
rnn_model.summary()
```
